| 기능             | `as_query_engine`                               | `as_retriever`                                 |
|------------------|-------------------------------------------------|------------------------------------------------|
| 목적             | 문서를 바탕으로 LLM 응답 생성                   | 문서 검색만 수행                                |
| 사용 방식        | 질문에 대한 완성된 답변을 제공하는 Q&A 시스템   | 검색된 문서를 그대로 반환하는 정보 검색 시스템 |
| 응답 방식        | 문서 검색 + 응답 생성                           | 문서 검색만 수행, LLM 응답 없음                |
| 주요 사용 사례   | RAG 기반 질문 응답 생성                         | 문서 검색 및 필터링                             |


## 예제 9.5 라마인덱스 내부에서 검색 증강 생성을 수행하는 과정
코드 출처: https://docs.llamaindex.ai/en/stable/understanding/querying/querying.html


- RAG 방식을 사용하여, 질문에 대해 가장 유사한 문서를 찾고 그 내용을 기반으로 답변을 생성
- RetrieverQueryEngine을 통해 검색 및 응답 생성을 통합하여 간편하게 질문에 대한 답변을 생성

```python
from llama_index.core import (
    VectorStoreIndex,
    get_response_synthesizer,
)
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.postprocessor import SimilarityPostprocessor

# 검색을 위한 Retriever 생성
retriever = VectorIndexRetriever(
    index=index,
    similarity_top_k=1,
)

# 검색 결과를 질문과 결합하는 synthesizer
response_synthesizer = get_response_synthesizer() # 검색된 결과를 기반으로 최종 응답을 생성

# 위의 두 요소를 결합해 쿼리 엔진 생성
query_engine = RetrieverQueryEngine(
    retriever=retriever,
    response_synthesizer=response_synthesizer,
    node_postprocessors=[SimilarityPostprocessor(similarity_cutoff=0.7)], # 이 값보다 유사도가 낮은 문서는 제외
)

# RAG 수행
response = query_engine.query("북태평양 기단과 오호츠크해 기단이 만나 국내에 머무르는 기간은?")
print(response)
# 장마전선에서 북태평양 기단과 오호츠크해 기단이 만나 국내에 머무르는 기간은 한 달 가량입니다.
```

## 예제 9.7 LLM 캐시를 사용하지 않았을 때 동일한 요청 처리에 걸린 시간 확인

```python
import time  # 시간 측정용 모듈 가져오기

# OpenAI 응답 객체에서 텍스트 응답만 추출하는 함수 정의
def response_text(openai_resp):
    return openai_resp.choices[0].message.content  # 응답 객체의 첫 번째 응답 메시지에서 텍스트 내용 반환

# 사용자가 묻고자 하는 질문
question = "북태평양 기단과 오호츠크해 기단이 만나 국내에 머무르는 기간은?"

# 두 번 반복하여 질문 및 응답 시간 측정
for _ in range(2):
    start_time = time.time()  # 시작 시간 기록

    # OpenAI API 요청 및 응답 받기
    response = openai_client.chat.completions.create(
      model='gpt-3.5-turbo',  # 사용할 모델 지정
      messages=[
        {
            'role': 'user',  # 사용자 역할 지정
            'content': question  # 사용자 질문 메시지 내용
        }
      ],
    )

    # 응답에서 텍스트 내용 추출
    response = response_text(response)

    # 질문 및 소요 시간 출력
    print(f'질문: {question}')
    print("소요 시간: {:.2f}s".format(time.time() - start_time))  # 응답 소요 시간 초 단위 출력
    print(f'답변: {response}\n')  # AI의 답변 출력

# 예시 출력:
# 질문: 북태평양 기단과 오호츠크해 기단이 만나 국내에 머무르는 기간은?
# 소요 시간: 2.71s
# 답변: 북태평양 기단과 오호츠크해 기단이 만나 국내에 머무르는 기간은 겨울 시즌인 11월부터 다음 해 3월까지입니다. 이 기간 동안 기온이 급격히 하락하며 한반도에 한기가 밀려오게 됩니다.

# 질문: 북태평양 기단과 오호츠크해 기단이 만나 국내에 머무르는 기간은?
# 소요 시간: 4.13s
# 답변: 북태평양 기단과 오호츠크해 기단은 겨울에 만나 국내에 머무르는 것이 일반적입니다. 이 기단들은 주로 11월부터 2월이나 3월까지 국내에 영향을 미치며, 한국의 겨울철 추위와 함께 한반도 전역에 형성되는 강한 서북풍과 냉기를 가져옵니다.
```

## 예제 9.9 유사 검색 캐시 추가 구현

**Semantic Cache**

요청과 유사한 의미(semantic similarity)를 가진 데이터를 캐싱하는 시스템

- 장점
  - 성능 최적화: 유사한 요청이 반복될 때 매번 새로운 응답을 생성하지 않고 캐시에서 응답을 반환하므로, 응답 속도가 빨라지고 자원이 절약
  - 유연한 캐싱: 단순한 텍스트 일치에 비해 의미적으로 유사한 요청도 처리할 수 있어, 다양한 표현의 요청에 대해 유사한 응답을 제공

```python
class OpenAICache:
    # 초기화 메서드, OpenAI 클라이언트와 유사 검색 캐시 저장
    def __init__(self, openai_client, semantic_cache):
        self.openai_client = openai_client  # OpenAI 클라이언트 저장
        self.cache = {}  # 일치 캐시 초기화
        self.semantic_cache = semantic_cache  # 유사 검색 캐시 저장

    # 프롬프트 생성 메서드
    def generate(self, prompt):
        # 일치 캐시 확인
        if prompt not in self.cache:
            # 유사 검색 캐시에서 유사한 문서 검색
            similar_doc = self.semantic_cache.query(query_texts=[prompt], n_results=1)
            # 유사도 기준 만족 여부 확인
            if len(similar_doc['distances'][0]) > 0 and similar_doc['distances'][0][0] < 0.2:
                return similar_doc['metadatas'][0][0]['response']  # 유사한 응답 반환
            else:
                # OpenAI 모델을 통한 새 응답 생성
                response = self.openai_client.chat.completions.create(
                    model='gpt-3.5-turbo',
                    messages=[
                        {
                            'role': 'user',
                            'content': prompt
                        }
                    ],
                )
                # 일치 캐시에 새 응답 저장
                self.cache[prompt] = response_text(response)
                # 유사 검색 캐시에 새 프롬프트 및 응답 저장
                self.semantic_cache.add(documents=[prompt], metadatas=[{"response": response_text(response)}], ids=[prompt])

        return self.cache[prompt]  # 캐시된 응답 반환
```

```python
#  대화의 흐름과 역할을 정의하는 텍스트
colang_content = """
define user greeting
    "안녕!"
    "How are you?"
    "What's up?"

define bot express greeting
    "안녕하세요!"

define bot offer help
    "어떤걸 도와드릴까요?"

define flow greeting
    user express greeting
    bot express greeting
    bot offer help
"""

# 사용할 모델의 설정을 YAML 형식으로 정의
yaml_content = """
models:
  - type: main
    engine: openai
    model: gpt-3.5-turbo

  - type: embeddings
    engine: openai
    model: text-embedding-ada-002
"""

# Rails 설정하기
## 봇이 사용자와 상호작용하는 대화 흐름과 모델 구성을 기반으로 Rails 시스템을 구성
config = RailsConfig.from_content(
    colang_content=colang_content,
    yaml_content=yaml_content
)
# Rails 생성
rails = LLMRails(config)

rails.generate(messages=[{"role": "user", "content": "안녕하세요!"}])
# {'role': 'assistant', 'content': '안녕하세요!\n어떤걸 도와드릴까요?'}
```

## 예제 9.16 OpenAI API 로깅하기

생성된 모델 응답과 각종 메타데이터를 추적할 수 있어, LLM의 성능 분석이나 디버깅에 활용

```python
import datetime  # 날짜와 시간을 다루기 위한 모듈 가져오기
from openai import OpenAI  # OpenAI API를 사용하기 위한 OpenAI 모듈 가져오기
from wandb.sdk.data_types.trace_tree import Trace  # W&B의 Trace 클래스를 가져오기


client = OpenAI()


system_message = "You are a helpful assistant."

query = "대한민국의 수도는 어디야?"

temperature = 0.2

model_name = "gpt-3.5-turbo"

# OpenAI API를 통해 응답 생성
response = client.chat.completions.create(
    model=model_name,  # 사용할 모델 지정
    messages=[{"role": "system", "content": system_message},
              {"role": "user", "content": query}],  # 시스템 메시지와 사용자 질문 설정
    temperature=temperature
)

# W&B의 Trace 객체 생성
root_span = Trace(
    name="root_span",  # 트레이스의 이름 지정
    kind="llm",  # 트레이스 종류 설정 ("llm" - 언어 모델 사용)
    status_code="success",  # 상태 코드 설정 (성공 여부 표시)
    status_message=None,  # 상태 메시지 (오류가 없으므로 None)
    metadata={
        "temperature": temperature,  # 메타데이터에 생성 온도 기록
        "token_usage": dict(response.usage),  # 응답의 토큰 사용량 기록
        "model_name": model_name  # 사용한 모델 이름 기록
    },
    inputs={
        "system_prompt": system_message,  # 입력으로 사용된 시스템 메시지 기록
        "query": query  # 입력으로 사용된 사용자 질문 기록
    },
    outputs={
        "response": response.choices[0].message.content  # 생성된 응답 내용 기록
    },
)

# 트레이스를 W&B에 로깅 (저장)
root_span.log(name="openai_trace")
```

## 예제 9.17 라마인덱스 W&B 로깅

```python
from datasets import load_dataset
import llama_index
from llama_index.core import Document, VectorStoreIndex, ServiceContext
from llama_index.llms.openai import OpenAI
from llama_index.core import set_global_handler

# OpenAI 모델 설정 및 W&B 로깅 핸들러 설정
llm = OpenAI(model="gpt-3.5-turbo", temperature=0)  # OpenAI의 GPT-3.5 Turbo 모델을 설정, temperature는 응답의 창의성 조절
set_global_handler("wandb", run_args={"project": "llamaindex"})  # W&B 로깅을 위한 글로벌 핸들러 설정, 프로젝트 이름은 "llamaindex"
wandb_callback = llama_index.core.global_handler  # 설정한 W&B 핸들러를 콜백으로 지정
service_context = ServiceContext.from_defaults(llm=llm)  # 서비스 컨텍스트 생성, LLM 모델 포함

# KLUE MRC 데이터셋 로드 및 문서 리스트 생성
dataset = load_dataset('klue', 'mrc', split='train')  # KLUE MRC 데이터셋의 train 분할을 로드
text_list = dataset[:100]['context']  # 처음 100개의 문서에서 'context' 텍스트 추출
documents = [Document(text=t) for t in text_list]  # 각 텍스트를 Document 객체로 변환하여 리스트에 저장

# 문서 인덱스 생성
index = VectorStoreIndex.from_documents(documents, service_context=service_context)  # 문서 리스트를 인덱싱하여 벡터 저장소 인덱스 생성

# 첫 번째 질문 출력
print(dataset[0]['question'])  # 첫 번째 질문 출력 ("북태평양 기단과 오호츠크해 기단이 만나 국내에 머무르는 기간은?" 출력 예상)

# 질의 엔진 설정
query_engine = index.as_query_engine(similarity_top_k=1, verbose=True)  # 인덱스를 질의 엔진으로 변환, 상위 1개의 유사 문서 반환하도록 설정, verbose=True로 상세 정보 출력

# 첫 번째 질문에 대한 질의 수행 및 응답 생성
response = query_engine.query(dataset[0]['question'])  # 첫 번째 질문에 대해 질의 수행하여 응답 생성
```
